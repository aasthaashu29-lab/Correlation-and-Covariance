# **Correlation and Covariance**

Q1. Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.

    -->Covariance: A measure of the joint variability of two random variables. It indicates the direction of the linear relationship.

Difference in Scale: Covariance can take any value from −∞ to +∞. Its units are the product of the units of X and Y. Correlation is a standardized version of covariance that always ranges from −1 to +1, making it dimensionless.

Interpretation: Because covariance is scale-dependent, you can't tell the "strength" of a relationship just by looking at the number. Correlation allows for an immediate assessment of strength regardless of the units used.

Q2. What does a positive, negative, and zero covariance indicate about the relationship between two variables?
   
    -->Positive: Both variables tend to move in the same direction (as X increases, Y increases).

Negative: Variables move in opposite directions (as X increases, Y decreases).

Zero: No linear relationship exists between the variables.

Q3. Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation
preferred in many cases?

    -->Limitations: The main limitation is scale sensitivity. If you change the unit of measurement (e.g., from meters to centimeters), the covariance value changes drastically, even though the relationship hasn't changed.

Preference: Correlation is preferred because it is unitless and standardized. It allows researchers to compare the strength of relationships across completely different datasets.

Q4. Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation
coefficient. When would you prefer to use Spearman’s correlation?

    -->Pearson’s (r): Measures the strength of a linear relationship between continuous variables. It assumes data is normally distributed.

Spearman’s (ρ): Measures the strength of a monotonic relationship using the ranks of the data.

When to use Spearman’s: Use it when your data is ordinal, when the relationship is non-linear (but still monotonic), or when your dataset contains significant outliers that would skew Pearson’s.


Q5. If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you
infer causation from this value? Why or why not?
  
    -->Interpretation: This indicates a strong, positive linear relationship.

Causation: No, you cannot infer causation. A high correlation only means the two variables move together; it doesn't prove that X causes Y. A third "lurking" variable could be influencing both.
